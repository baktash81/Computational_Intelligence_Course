```python
import pandas as pd

data1 = pd.read_csv('./data1.csv')
data1
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Normalize and standardize the X and Y columns
data1[['X', 'Y']] = scaler.fit_transform(data1[['X', 'Y']])

data1
```

```python
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Define the range of c values
c_values = range(2, 11)

# Iterate over each c value
for c in c_values:
    # Apply FCM clustering algorithm
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(data1.T, c, 2, error=0.005, maxiter=1000)

    # Assign each datapoint to the cluster with the highest membership value
    cluster_membership = u.argmax(axis=0)

    fig, ax = plt.subplots()
    
    for i in range(c):
        ax.plot(data1['X'][cluster_membership == i], data1['Y'][cluster_membership == i], 'o')

    for cluster_center in cntr:
        ax.plot(cluster_center[0], cluster_center[1], 'X', markersize=10, markeredgewidth=2)


    # Plot the graph with different colors for each cluster
    ax.set_title(f'Fuzzy C-Means Clustering with C={c}')
    ax.legend()
    plt.show()
```

# Best FPC

```python
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Define the range of c values
c_values = range(2, 11)

# Initialize variables to store the best c value and the corresponding FPC score
best_c = None
best_fpc = -1
best_values = []

# Iterate over each c value
for c in c_values:
    # Apply FCM clustering algorithm
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(data1.T, c, 2, error=0.005, maxiter=1000)

    # Check if the current FPC score is higher than the previous best score
    if fpc > best_fpc:
        best_c = c
        best_fpc = fpc
        best_values = [cntr, u, u0, d, jm, p, fpc]

    
# Assign each datapoint to the cluster with the highest membership value
cluster_membership = best_values[1].argmax(axis=0)

fig, ax = plt.subplots()

for i in range(c):
    ax.plot(data1['X'][cluster_membership == i], data1['Y'][cluster_membership == i], 'o')

for cluster_center in best_values[0]:
    ax.plot(cluster_center[0], cluster_center[1], 'X', markersize=10, markeredgewidth=2)

# Plot the graph with different colors for each cluster
ax.set_title(f'Fuzzy C-Means Clustering with C={c}')
ax.legend()
plt.show()

# Print the best value of c and the corresponding FPC score
print(f"The best value of c is {best_c} with an FPC score of {best_fpc}")
```
